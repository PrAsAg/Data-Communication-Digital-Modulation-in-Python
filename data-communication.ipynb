```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def gen_sin():
    t=np.linspace(0,2*np.pi,1000)
    y=np.sin(2*np.pi*t)
    plt.figure(figsize=(100,10))
    plt.subplot(231)
    plt.plot(t,y)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Sinusoidal Wave')
    plt.show()
```

```python
gen_sin()
```

```python
def gen_cos():
    t=np.linspace(0,2*np.pi,1000)
    y=np.cos(2*np.pi*t)
    plt.figure(figsize=(100,10))
    plt.subplot(231)
    plt.plot(t,y,'r')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Cosine Wave')
    plt.show()
```

```python
gen_cos()
```

```python
def gen_unit_step():
    n=10
    t=np.arange(10)
    y=np.ones(10)
    plt.figure(figsize=(100,10))
    plt.subplot(231)
    plt.stem(t,y,'y')
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.title('Unit Step Funtion')
    plt.show()
```

```python
gen_unit_step()
```

```python
def gen_exp():
    t= np.linspace(0,20,1000)
    #give amplitude
    a= 0.5
    y=np.exp(a*t)
    plt.figure(figsize=(100,25))
    plt.subplot(231)
    plt.plot(t,y,'r--')
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.title('Exponential Signal')
    plt.show()
```

```python
gen_exp()
```

```python
def gen_unit_impulse():
    t=np.linspace(-10,10,21)
    y=[1 if i == 0 else 0 for i in t]
    plt.figure(figsize=(100,10))
    plt.subplot(231)
    plt.stem(t,y)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Unit Impulse Signal')
    plt.show()
    
```

```python
gen_unit_impulse()
```

```python
def gen_ramp():
    t=np.arange(50)
    plt.figure(figsize=(100,10))
    plt.subplot(231)
    plt.stem(t,t)
    plt.xlabel('Sequence')
    plt.ylabel('Amplitude')
    plt.title('Ramp Signal')
    plt.show()
```

```python
def PSK():
    fc= 100#input('Enter Frequency of Sin Wave Carrier:')
    fp=5#input('Enter the freq of Periodic Binary pulse (Message):')
    amp=1#input('Enter the amplitude (For Carrier & Binary Pulse Message):')
    t=np.linspace(0,1,10000)
    c= float(amp)*np.sin(2*np.pi*fc*t)
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    plt.plot(t,c)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Carrier Wave')
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    m=signal.square(2 * np.pi * fp * t)
    plt.plot(t,m)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Binary Message Pulses')
    w=c*m
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    plt.plot(t,w)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Phase Shift Keying Signal')
```

```python
PSK()
```

```python
def ASK():
    fc= 100#input('Enter Frequency of Sin Wave Carrier:')
    fp=5#input('Enter the freq of Periodic Binary pulse (Message):')
    amp=1#input('Enter the amplitude (For Carrier & Binary Pulse Message):')
    t=np.linspace(0,1,10000)
    c= float(amp)*np.sin(2*np.pi*fc*t)
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    plt.plot(t,c)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Carrier Wave')
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    m=(amp/2)*signal.square(2*np.pi*fp*t)+(amp/2)
    plt.plot(t,m)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Binary Message Pulses')
    w=c*m
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    plt.plot(t,w)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Amplitude Shift Keying Signal')
```

```python
ASK()
```

```python
def FSK():
    fc= 100#input('Enter Frequency of Sin Wave Carrier:')
    fp=5#input('Enter the freq of Periodic Binary pulse (Message):')
    amp=1#input('Enter the amplitude (For Carrier & Binary Pulse Message):')
    t=np.linspace(0,1,10000)
    c= float(amp)*np.sin(2*np.pi*fc*t)
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    plt.plot(t,c)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Carrier Wave')
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    m=(amp/2)*signal.square(2*np.pi*fp*t)+(amp/2)
    plt.plot(t,m)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Binary Message Pulses')
    f= fc+fc*m/2
    fsk=np.sin(2*np.pi*f*t)
    plt.figure(figsize=(100,5))
    plt.subplot(231)
    plt.plot(t,fsk)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Frequency Shift Keying Signal')
    
```
